```python
import torch
from matplotlib import pyplot as plt 
import numpy as np 
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5','yolov5m')
```

```python
imgs = ['https://upload.wikimedia.org/wikipedia/commons/thumb/e/e4/Cars_in_traffic_in_Auckland%2C_New_Zealand_-_copyright-free_photo_released_to_public_domain.jpg/800px-Cars_in_traffic_in_Auckland%2C_New_Zealand_-_copyright-free_photo_released_to_public_domain.jpg?20190503075536']
results = model(imgs)
```

```python
results
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(2)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
import uuid
import time

# Define the absolute path to save images
IMAGES_PATH = '/home/p4nda/Documents/yolo/yolov5/data/images'
os.makedirs(IMAGES_PATH, exist_ok=True)

labels = ['awake', 'drowsy']  # Replace with your actual labels
number_imgs = 20  # Number of images to capture per label

# Open the camera at index 2 (change if needed)
cap = cv2.VideoCapture(2)
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

time.sleep(2)  # Allow the camera to warm up

# Loop through labels
for label in labels:
    print(f'Collecting images for {label}')
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print(f'Collecting images for {label}, image number {img_num}')
        
        # Webcam feed
        ret, frame = cap.read()
        if not ret or frame is None or frame.size == 0:
            print(f"Failed to capture frame for {label}, image number {img_num}")
            continue
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, f'{label}.{uuid.uuid1()}.jpg')
        
        # Writes out image to file 
        success = cv2.imwrite(imgname, frame)
        if not success:
            print(f"Failed to save image {imgname}")
        else:
            print(f"Image saved successfully: {imgname}")
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

# Define the path to save the image
IMAGES_PATH = os.path.expanduser('~/images')
os.makedirs(IMAGES_PATH, exist_ok=True)

# Open the camera at index 0 (change to 1 or 2 if needed)
cap = cv2.VideoCapture(2)
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

# Capture a single frame
ret, frame = cap.read()
if ret and frame is not None and frame.size > 0:
    # Define the image filename and path
    imgname = os.path.join(IMAGES_PATH, 'test_image.jpg')
    
    # Save the captured frame as an image
    success = cv2.imwrite(imgname, frame)
    if success:
        print(f"Image saved successfully at {imgname}")
    else:
        print(f"Failed to save image at {imgname}")
else:
    print("Failed to capture a valid frame.")

# Release the camera and close any open windows
cap.release()
cv2.destroyAllWindows()
```
